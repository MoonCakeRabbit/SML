```python
file_path = 'C:\\Users\\jingjun\\Desktop\\sml data\\set2.test.txt'

with open(file_path, 'r') as file:
    #读取文件中的所有行，将每行作为一个字符串元素存储在 lines 列表中。
    lines = file.readlines()
```

```python
import re

#创建一个空列表 data 用于存储解析后的数据。
data = []
for line in lines:
    # label 是每行的第一个元素，转换为整数。
    label = int(line.split()[0])
    #qid 是查询 ID，使用正则表达式 r'qid:(\d+)' 提取。如果匹配成功，则转换为整数；否则设置为 None。
    qid_match = re.search(r'qid:(\d+)', line)
    qid = int(qid_match.group(1)) if qid_match else None

    #将每条记录存储为字典 {'label': label, 'qid': qid, 'features': features_dict} 并添加到 data 列表中。
    # features 使用正则表达式 r'(\d+):([\d.]+)' 提取特征索引和特征值。
    features = re.findall(r'(\d+):([\d.]+)', line)
    # features_dict 将提取到的特征对 (index, value) 转换为字典格式，索引作为键，值作为浮点数
    features_dict = {int(index): float(value) for index, value in features}
    
    # 保存数据
    data.append({'label': label, 'qid': qid, 'features': features_dict})

print(data[:5])  # 查看前5条记录
```

```python
import pandas as pd

# 构建 DataFrame
records = []
for entry in data:
    record = {'label': entry['label'], 'qid': entry['qid']}
    record.update(entry['features'])
    records.append(record)

df = pd.DataFrame(records)
print(df.head())
```

```python
# 检查缺失值
print(df.isna().sum())

# 填补缺失值，例如用0填充
df.fillna(0, inplace=True)
```
